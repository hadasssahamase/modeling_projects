```python
# Loading the dataset
import pandas as pd
from surprise import Dataset, Reader

# Load the dataset (MovieLens 100k)
df = pd.read_csv("https://files.grouplens.org/datasets/movielens/ml-100k/u.data", 
                 sep="\t", names=["user_id", "item_id", "rating", "timestamp"])

df = df.drop(columns=["timestamp"])  # Drop timestamp column

# Load movie names
movies = pd.read_csv("https://files.grouplens.org/datasets/movielens/ml-100k/u.item", 
                     sep="|", encoding="latin-1", header=None, usecols=[0,1], names=["item_id", "title"])

# Merge movie titles with ratings
df = df.merge(movies, on="item_id")

print(df.head())  # Preview dataset
```
